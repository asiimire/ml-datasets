```python
"""
Problem Statement

You have been provided with a dataset that lists Ohio States leading causes of death from the year 2012.
Using the two data points cause of death and percentile, draw a pie chart to visualize the dataset
"""
```

```python
#import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Data from Ohio state, leading causes of death in the state from year 2012
#cause of death
cause = 'Chronic diseases','Unintentional injuries','Alzhemiers','Influenza and Pneumonia','Sepsis','Others'
#percentile
percentile = [62,5,4,2,1,26]
```

```python
#set the piechart plot properties
#set the figure size
plt.figure(figsize=(10,10))

#explode the largest pie in the dataset(it slices out the largest piece based on the percentile)
explode = (0.05,0,0,0,0,0)

#set the piechart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%', startangle=70)
#autopct will embed the percntile vlaue to each slice of the pie

#set axis equal, to draw pie as circle
plt.axis('equal')
#set the title of the piechart
plt.title('Ohio State - 2012 : Leading Causes of Death')
#show the plot(piechart)
plt.show()
```

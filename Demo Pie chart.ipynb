```python
#mport the required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#job data in percentile
job_data = ['48','28','17','8','5','10']

#define the labels as different departments
labels = 'IT','Finance','Marketing','Admin','HR','Operations'

#explode the first slice which is IT
explode = [0.05,0,0,0,0,0]

#draw the piechart and set the parameters
plt.pie(job_data,labels=labels,explode=explode)

#show the plot
plt.show()
```

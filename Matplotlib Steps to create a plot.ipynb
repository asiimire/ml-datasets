```python
#import numpy for generating random numbers
import numpy as np
#import matplotlib library
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
"""
Numpy is used to generate the random numbers
Pyplot,which is built-in python library is used to plot numbers
style class is used for setting the grid style
%matplotlib inline is required to display the plot within jupyter notebook
"""
```

```python
#generate random numbers (total 10)
randomNumber = np.random.rand(10)
```

```python
#view them
print(randomNumber)
```

```python
#select the style of the plot
style.use('ggplot')
#plot the random number
plt.plot(randomNumber, 'g' ,label='line one', linewidth=2)
#x-axis is the number of random numbers (index)
plt.xlabel('Range')
#y-axis is the actual random number
plt.ylabel('Numbers')
plt.title('First plot')

plt.legend()
plt.show()

"""
legend method is used to plot the graph based on the set conditions
show method is used to display the created plot
"""
```
